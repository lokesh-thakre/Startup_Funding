```python
most_fund = df.groupby(['Starting Year'], as_index =False)['Funding Amount in $'].sum().sort_values(by='Funding Amount in $', ascending = False)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x = 'Starting Year',y = 'Funding Amount in $', data = most_fund)
```

```python
dict = {'City' : ['Mumbai', 'Bengaluru', 'Pune', 'Noida', 'New Delhi']}
df = pd.DataFrame(dict)
city_counts = df['City'].value_counts()
city_counts.plot(kind ='pie', color='skyblue')

plt.xlabel('City')
plt.ylabel('Number of Startups')
  # Rotate x-axis labels for better readability
plt.show()
```
